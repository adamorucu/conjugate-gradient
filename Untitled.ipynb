```python
import matplotlib.pyplot as plt
from matplotlib import ticker
import numpy as np
import seaborn; seaborn.set()
```

```python
class Roofline:
    def __init__(self, cpu_speed, bandwith):
        self.cpu = cpu_speed
        self.bandwith = bandwith
        
    def vis(self, I=None, legend=[]):
        fig, ax = plt.subplots()
        arr = np.arange(1/16, 16, 1/16)
        ax.plot(arr, [arr[i]*self.bandwith for i in range(len(arr))], 'g')
        ax.plot(arr, len(arr)*[self.cpu], 'g')
        if I != None:
            for i, f in I.items():
                peak = min(self.bandwith*i, self.cpu)
                print(f'Peak for {i} is at {peak}')
#                 ax.scatter([i], [f], color='red', linewidths=2)
                ax.plot([i,i], [0, peak], 'r--', linewidth=0.7)
        
        plt.title("Roofline model")
        plt.xlabel("Arithmetic Intensity (FLOPS/byte)")
        plt.ylabel("Achievable GFLOPS/sec")
        ax.set_xscale("log")
        ax.set_yscale("log")
        ax.set_xticks([1/16, 1/8,1/4,1/2,1,2,4,8,16])
        ax.set_yticks([2,4,8,16,32,64,128,256,512])
        ax.get_xaxis().set_major_formatter(ticker.ScalarFormatter())
        ax.get_yaxis().set_major_formatter(ticker.ScalarFormatter())
        plt.show()
#         plt.savefig('img/roofline.png')
```
